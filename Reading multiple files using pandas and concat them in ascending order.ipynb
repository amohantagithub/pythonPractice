## Reading multiple excel files using pandas and concatenating them sequentially   

```python
# This works with Python 3.6
# Run the  script in the directory which contains the excel files.
# In my directory, the filenames of the excelfiles are "file1.xlsx", "file2.xlsx", etc. First six rows     doesnot contain data.
# The following code reads the excel file and concat them in order.

import pandas as pd 
import glob
```

```python
file1=pd.read_excel("file1.xlsx", skiprows=6, index_col=0) #Reading the first file to look at the structure of the data.  
file1
```

```python
filenames=glob.glob("*.xlsx")   # This will give a list of files without any order.
print(filenames)
```

```python
filenames_order=["file{}.xlsx".format(f) for f in range(len(filenames)) if "file{}.xlsx".format(f) in filenames]
print(filenames_order)
```

```python
#Reading the excel files in order  
df_files=[pd.read_excel("file{}.xlsx".format(f), skiprows=6, index_col=0) for f in range(len(filenames)) if "file{}.xlsx".format(f) in filenames]
df_files[0]   
```

```python
dfc=pd.concat(df_files, keys=filenames_order) 
dfc.head()
```

```python
dfs=dfc.unstack()
dfs.head(9)
```
